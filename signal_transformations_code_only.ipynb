```
import numpy as np
import matplotlib.pyplot as plt

# =============================
# سیگنال‌های گسسته (n)
# =============================

n = np.arange(-10, 11, 1)
x = lambda n: np.sin(0.5 * n)

x1 = x(n - 2)
x2 = x((3/2) * n + 3)
x3 = x(n) + x(2 * n) - 3

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.stem(n, x1, basefmt=" ", use_line_collection=True)
plt.title("x[n - 2]")
plt.xlabel("n"); plt.ylabel("Amplitude")

plt.subplot(1, 3, 2)
plt.stem(n, x2, basefmt=" ", use_line_collection=True)
plt.title("x[(3/2)n + 3]")
plt.xlabel("n"); plt.ylabel("Amplitude")

plt.subplot(1, 3, 3)
plt.stem(n, x3, basefmt=" ", use_line_collection=True)
plt.title("x[n] + x[2n] - 3")
plt.xlabel("n"); plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()

# =============================
# سیگنال‌های پیوسته (t)
# =============================

t = np.linspace(-10, 10, 1000)
x_c = lambda t: np.cos(2 * np.pi * 0.2 * t)

x4 = x_c(3 * t - 0.5)
x5 = x_c(0.5 * t + 3)
x6 = x_c(t - 2) - t

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.plot(t, x4)
plt.title("x(3t - 1/2)")
plt.xlabel("t"); plt.ylabel("Amplitude")

plt.subplot(1, 3, 2)
plt.plot(t, x5)
plt.title("x(1/2 t + 3)")
plt.xlabel("t"); plt.ylabel("Amplitude")

plt.subplot(1, 3, 3)
plt.plot(t, x6)
plt.title("x(t - 2) - t")
plt.xlabel("t"); plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
